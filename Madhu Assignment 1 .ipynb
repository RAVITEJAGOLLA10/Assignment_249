```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sales = pd.read_csv(r"C:\Users\65866\Downloads\2357220-Visualization_249_26102023_-_Suresh_Nagulavancha (1)\Sales_Data.csv")
```

## Line Plot using Pandas

```python
df_sales[["Unit_MakingCost","Total_Cost"]].plot(x="Unit_MakingCost", y="Total_Cost", c="b", lw=5, marker="*", mfc="r", mec="c", ms=10)
plt.title("Unit_MakingCost vs Total_Cost")
plt.show()
```

```python
## The above lineplot shows that comparision between Unit_MakingCost and Total_Cost.We can see that as the graph increases Unit_MakingCost gradually the Total_Cost also increases
## From the graph i can conclude that the Maximum Total_Cost increases with increase in Unit_MakingCost
```

## Line Plot using Matplotlib

```python
df_sales.columns
```

```python
plt.plot(df_sales["Units_Sold"].head(30), df_sales["Unit_SellingPrice"].head(30), label="Units",marker="P",mfc="r",mec="y",ms=10,lw=2)
plt.title("Units_Sold vs Unit_SellingPrice", loc="center",fontdict={'family':'sans','style':'italic','color':'blue', 'size':18})
plt.xlabel("Units_Sold")
plt.ylabel("Unit_SellingPrice")
plt.legend()
plt.grid()
plt.show()                                                       
```

```python
# From above plot i conclude that most Units Sold happened from 2000-10000 units, however the highest selling price unit is around 620
```

## Scatter Plot using Matplotlib

```python
plt.scatter(df_sales["Total_Revenue"],df_sales["Total_Cost"], label="Total_Revenue vs Total_Cost", edgecolors="y",marker="p")
plt.title("Total_Revenue vs Total_Cost")
plt.xlabel("Total_Revenue")
plt.ylabel("Total_Cost")
plt.legend()
plt.grid()
plt.show()
```

```python
# From above scatter plot i can conclude that as the Total_Cost increases gradually the Total_Revenue also increases
```

## Histogram using Pandas
